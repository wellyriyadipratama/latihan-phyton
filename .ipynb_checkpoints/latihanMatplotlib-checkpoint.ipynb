```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#membuat data array di mulai dari angka 1 sampai 5dengan jumlah index ada 11
x = np.linspace(1,5,11)
#dari array yang telah di buat di atas di beri pangkat 2
y = x**2
```

```python
x
```

```python
y
```

```python
#membuat plot dari data x dan y
plt.plot(x, y, 'r') #r merupakan warna yaitu red atau merah
plt.xlabel('X Axis') #label untuk sumbu X
plt.ylabel('Y Axis') #label untuk sumbu Y
plt.title('Data Contoh') #label untuk judul diagram
plt.show() #untuk memunculkan diagramnya
```

```python
#membuat bnyk diagram atau plot di 1 canvas
plt.subplot(1,2,1) #1 adalah row #2 adalah kolom #1 adalah plot
plt.plot(x,y, 'r--') #r-- adalah garis berwarna merah putus-putus
plt.subplot(1,2,2) #1 adalah row #2 adalah kolom #2 adalah plot
plt.plot(y,x, 'b*-') #b* adalah garis berwarna biru dengan campuran bintang dan garis
```

```python
#membuat canvas kosong 
fig = plt.figure()
#mengatur ukuran kanvas
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
#0.1 jarak dari kiri
#0.1 jarak dari bawah
#0.8 lebar
#0.8 tinggi

axes.plot(x, y, 'b')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Contoh Judul')
```

```python
#Plot di dalam plot
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #plot utama
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) #plot didalam

#pengaturan plot utama
axes1.plot(x, y, 'b')
axes1.set_xlabel('X Utama')
axes1.set_ylabel('Y Utama')
axes1.set_title('Plot Utama')

#pengaturan plot dalam
axes2.plot(x, y, 'r')
axes2.set_xlabel('X dalam')
axes2.set_ylabel('Y dalam')
axes2.set_title('Plot Dalam');
```

```python
#mengatur canvas
#figsize adalah pengaturan ukuran terhadap canvas yang akan di buat
#dpi adalah bentukan dari suatu ukuran hasil cetak atau visual semakin tinggi maka semakin besar gambarnya
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(8,4))

axes.plot(x, y, 'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title');
```

```python
#save atau export figure
fig, axes = plt.subplots(figsize=(8,4))

axes.plot(x, y, 'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('title');

#save atau export figure menjadi format PNG, JPG, EPS, SVG, PGF dan PDF.
fig.savefig("contoh.png", dpi=200)
```

```python
#menambahkan informasi ke dalam plot
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="data x pangkat 2")
ax.plot(x, x**3, label="data x pangkat 3")

#menambahkan info ke dalam figure
ax.legend(loc=0)
#loc = posisi dimana kita akan menaruhnya
#0 default kiri atas
#1 = kanan atas
#2 = kiri atas
#3 = kiri bawah
#4 = kanan bawah
```

```python
#style plot
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x+5, color="green", lw=3, linestyle='-') #bisa kita singkat
ax.plot(x, x+6, color="green", lw=3, ls='-.') #linewidth|
ax.plot(x, x+7, color="green", lw=3, ls=':') #linestyle
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5,5,15,5]) #format (panjang ls, jarak ls, panjang garis, jarak garis)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x+9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x+13, color="purple", lw=3, ls='-', marker='o', markersize=10)
ax.plot(x, x+14, color="purple", lw=3, ls='--', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=3, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=3, ls='--', marker='s', markersize=8,
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
```
