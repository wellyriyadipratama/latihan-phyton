```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
for link in soup.find_all('div',class_="ta-product"):
    dataImg = link.find('img', class_="entered")
    #print(dataImg['src']
    
    name=link.find('span', class_="ta-product-title")
    #print(name.get_text())
    
    price=link.find('span', class_="ta-product-price")
    print(price.get_text())
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
for link in soup.find_all('div',class_="ta-product"):
    dataImg = link.find('img', class_="entered")
    #print(dataImg['src']
    
    name=link.find('span', class_="ta-product-title")
    #print(name.get_text())
    
    price=link.find('span', class_="ta-product-price")
    #print(price.get_text())
    
    products.append(name.get_text())
    prices.append(name.get_text())
    image.append(dataImg['src'])
```

```python
#dari data di atas kita simpan ke dalam bentuk csv
df = pd.DataFrame({'Products Name':products,'Price':prices, 'Image':image})
df.to_csv('products.csv', index=False, encoding='utf-8')
```
