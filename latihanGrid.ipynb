```python
import seaborn as sns
import matplotlib.pyplot as plt
#membuat data dummy iris
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
#memvisualisasikan data dengan pairgrid
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
#map untuk mengisi data grid dengan plot
#mengisi pair grid dengan plot scatter
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
#memvisualisasikan data dengan pairplot
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species',palette='rainbow')
```

```python
tips = sns.load_dataset('tips')
```

```python
#memvisualisasikan data dengan facetgrid
#dengan kolom time dan row adalah smoker
#membuat data siapa saja yang menolak dan siapa saja yang tdk merokok beserta waktunya
g = sns.FacetGrid(tips, col="time",row="smoker")
```

```python
#dari grid sebelumnya kita buat histogramnya
g = sns.FacetGrid(tips, col="time",row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
#facetGrid
g = sns.FacetGrid(tips, col="time", row="smoker", hue="sex")
#add_lagend menambahkan info di samping
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```

```python
#memvisualisasikan data dengan joint grid atau untuk joint plot
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.distplot)
```

```python
#Regregasi
tips = sns.load_dataset('tips')
tips.head()
```

```python
#memvisualisasikan data dengan lmplot
sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex")
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", palette="coolwarm")
```

```python
#mengubah simbol hue dengan marker
#Scatter_kwa adalah besar markernya
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", palette="coolwarm",markers=['o','v'],scatter_kws={'s':1000})
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex")
```

```python
sns.lmplot(x="total_bill", y="tip",row='sex', data=tips, col="time")
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", col="day", palette="coolwarm")
```

```python
#aspect adalah lebar grid
#height adalah tinggi grid
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", col="day", palette="coolwarm", aspect=0.6,height=8)
```
