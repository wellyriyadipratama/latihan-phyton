```python
import pandas as pd
```

```python
#buat Dataframe
data = {
    'Perusahaan':['Babastudio','Babastudio','Techfor','Techfor'],
    'Nama':['Nur','Reza','Gusti','Pandu'],
    'Penjualan':[245,250,350,400]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
#menggabukngkan data ketika ada data yng sama 
```

```python
#membuat group untuk kolom perusahaan
df.groupby('Perusahaan')
```

```python
#menampilkan rata-rata penjualan dari masing-masing perusahaan
df.groupby('Perusahaan').mean()
```

```python
#menampilkan data terkecil dari masing2 perusahaan
df.groupby('Perusahaan').min()
```

```python
#menampilkan data tebesar dari masing2 perusahaan
df.groupby('Perusahaan').max()
```

```python
#menampilkan jumlah record di setiap perusahaan
df.groupby('Perusahaan').count()
```

```python
#menampilkan data data group
df.groupby('Perusahaan').describe()
```

```python
#menampilkan data data group secara horizontal
df.groupby('Perusahaan').describe().transpose()
```

```python
#menampilkan data2 perusahaan babastudio
df.groupby('Perusahaan').describe().transpose()['Babastudio']
```

```python
#MARGE
#menggabungkan 2 dataframe menjadi 1
```

```python
left = pd.DataFrame({
    'key1':['K0','K0','K1','K2'],
    'key2':['K0','K1','K0','K1'],
       'A':['A0','A1','A2','A3'],
       'B':['B0','B1','B2','B3'],})

right = pd.DataFrame({
    'key1':['K0','K1','K1','K2'],
    'key2':['K0','K0','K0','K0'],
       'C':['C0','C1','C2','C3'],
       'D':['D0','D1','D2','D3'],})
```

```python
left
```

```python
right
```

```python
#How = Inner (tipe merger)
#Dimana ketika tidak ada yg cocok antara key1 dan 2 maka dia tidak akan tampil
pd.merge(left, right, on=['key1','key2'])
```

```python
#how = outer
#dimana data akan di tampilkan semua ketika ada data yang tidak cocok maka akan di tulis NaN
pd.merge(left, right, how='outer', on=['key1','key2'])
```

```python
#how = right
#dimana dataframe right akan di tampilkan semua dan dataframe left akan di tampilkan sesuai dengan data right
pd.merge(left, right, how='right', on=['key1','key2'])

# Note :
# 1. Right key1 & key2 index 0 sama dengan left key1 & key2 index 0 Menjadikan A dan B ditampilkan
# 2. Right key1 & key2 index 1 sama dengan left key1 & key2 index 2 Menjadikan A dan B ditampilkan
# 3. Right key1 & key2 index 2 sama dengan left key1 & key2 index 2 Menjadikan A dan B ditampilkan
# 4. Right key1 & key2 index 3 tidak sama dengan left key1 & key2 di index manapun menjadikan nilai A dan B menjadi NaN
```

```python
#how = left
#dimana dataframe left akan di tampilkan semua dan dataframe right akan di tampilkan sesuai dengan data right
pd.merge(left, right, how='left', on=['key1','key2'])

# Note :
# 1. Left key1 & key2 index 0 sama dengan right key1 & key2 index 0 Menjadikan C dan D ditampilkan
# 2. Left key1 & key2 index 1 tidak sama dengan right key1 & key2 di index manapun menjadikan nilai C dan D menjadi NaN
# 3. Left key1 & key2 index 2 sama dengan right key1 & key2 index 1 dan index 2 menjadikan dia di tampilkan 2 data Menjadikan C dan D ditampilkan
# 4. Left key1 & key2 index 3 tidak sama dengan right key1 & key2 di index manapun menjadikan nilai C dan D menjadi NaN
```

```python
#cara lain dari pandas untuk menggabungkan 2 dataframe
left = pd.DataFrame({
       'A':['A0','A1','A2'],
       'B':['B0','B1','B2']},
        index=['K0','K1','K2'])

right = pd.DataFrame({
       'C':['C0','C1','C2'],
       'D':['D0','D1','D2']},
        index=['K0','K2','K3'])
```

```python
left.join(right, how='outer')
```

```python
left.join(right, how='left')
```
