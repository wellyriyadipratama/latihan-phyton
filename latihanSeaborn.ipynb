```python
import seaborn as sns
```

```python
#dummy data seaborn
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#DISPLOT
#memvisualisasikan data ke bentuk histogram
sns.distplot(tips['total_bill'])
```

```python
#kde false menghilangkan garis atau kernel pada plot
#bins jumlah batang yang ingin di tampilkan
sns.distplot(tips['total_bill'],kde=False,bins=10)
```

```python
#JOINPLOT 
#Membuat diagram atau plot scatter
#dimana datanya adalah variable tips
#data x nya total_bill atau bisa liat histogramnya di paling atas
#data y nya adalah tips atau bisa liat histogramnya di sebelah kiri
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
#membuat diagram atau plot hex
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
#Membuat diagram atau plot reg
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#menampilkan beberapa plot yng saling berhubungan
sns.pairplot(tips)
```

```python
#pairplot
#hue adalah pemberian warna yang berbeda pada suatu data
#palette adalah warna dari hue
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
#hanya menggambar tanda hubung untuk setiap titik pada ditribusi
sns.rugplot(tips['total_bill'])
```

```python
#heatmap
sns.heatmap(tips.corr())
```

```python
#cmap jenis pewarnaan
#annot menampilkan nilainya
sns.heatmap(tips.corr(), cmap="coolwarm",annot=True)
```

```python
#membuat pivot table
#yang isinya adalah passangers
#indexnya month
#kolomnya years
flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
data = flights.pivot_table(values='passengers',index='month',columns='year')
sns.heatmap(data)
```

```python
#custom headmap
#linecolor adalah warna bari pemisah data 
#linewidths adalah lebar baris pemisah data
sns.heatmap(data,cmap='magma',linecolor='white',linewidths=1)
```

```python
#clustermap
#memvisualisasikan data berdasarkan clustermap
sns.clustermap(data)
```

```python
#mengurutkan data
sns.clustermap(data, cmap='coolwarm', standard_scale=1)
```
